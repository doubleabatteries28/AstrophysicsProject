```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
m = np.linspace(0, 15, 100)
d = 0.2

def flux(f_o, m):
    return f_o * 10**(-m/2.5)

def area (d):
    return (((d/2)**2 * np.pi) - ((0.1/2)**2 * np.pi))

def time(f_o, m, d, wave):
    return ((3000 * 27) / (0.6*1.51*10**7* flux(f_o, m) * wave * area(d))) 
            #multiple of 27 comes from gaussain so peak is at 3000 photons
            #0.6 comes from the sesnitivity of the cammera, it only measures 60% of photons are measured
```

```python
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(1,1,1)
ax.set_yscale('log')

ax.plot(m, time(4260 ,m, d, 0.22,), color = 'blue', label = 'Blue filter')  #Blue filter
ax.plot(m, time(3080, m, d, 0.23,), color = 'red', label = 'Red filter' )  #Red filter
ax.plot(m, time(3730 ,m, d, 0.14,), color = 'green', label = 'Blue filter')  #Green filter
#ax.plot(m, time(?? ,m, d, ??,), color = 'black')  #Luminance filter - unknown values

ax.plot(6.4, 0.6, marker='o', linestyle='none', color = 'black', label = 'RZ Cas')
ax.plot(9.48, 30, marker='s', linestyle='none', color = 'black', label = 'RZ Cep')
ax.plot(10.97, 60, marker='^', linestyle='none', color = 'black', label = 'RV Uma')

ax.legend()

plt.rcParams.update({'font.size':18})

ax.set_xlabel('Apparent mag / Wm^-2',)
ax.set_ylabel('Time / s')
```
